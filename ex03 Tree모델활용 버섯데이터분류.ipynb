### 복습진행시 여러컬럼 확인해보기

```python
# 모델링
# 1. 현재 데이터는 문자형 - 수치 계산을 통해서 예측 불가능
#    문자 데이터를 수치로 바꿔주는 인코딩 작업
# 2. 훈련 / 평가 데이터 분할
# 3. 모델 불러오기 및 학습 진행 -> 성능 평가
# 4. 교차검증
# 5. 특성 중요도 확인 -> 결과 리포트
```

```python
# 인코딩 방법 - 2가지 방법
# 1. Label Encoding
# 2. one-hot Encoding(더미변수화) - 채택
```

```python
# 라벨 인코딩 실습
# sklearn.preprocessing : 전처리 도구 모음집
from sklearn.preprocessing import LabelEncoder
```

```python
labeler = LabelEncoder() # 객체 생성
labeler.fit(data['poisonous']) # 라벨 인코딩 변환패턴 학습
label = labeler.transform(data['poisonous']) # 실제 데이터 변환
label
```

```python
labeler.classes_
```

```python
labeler.inverse_transform(label)
```

```python
# 원 핫 인코딩 진행
# 우리는 훈련/평가 데이터 분할 되기전 인코딩 진행했다.
# X_oh = pd.get_dummies(X) -> True/False로 값 나옴 -> 컴퓨터가 인식못함
X_oh = pd.get_dummies(X, dtype = int)
X_oh
```

### 인코딩 시 주의점
- 훈련 데이터와 평가 데이터로 나눠진 경우 각각 데이터에 동일한 방식으로 인코딩을 진행

```python
# 훈련 평가 데이터 분할하기
# train_test_split 함수 사용하기
from sklearn.model_selection import train_test_split
```

```python
# 인코딩 시킨 데이터를 쪼개야 된다. -> X_oh (원 핫 인코딩 진행)
X_train, X_test, y_train, y_test = train_test_split(X_oh,
                                                    y,
                                                    random_state = 10,
                                                    test_size = 0.3,
                                                    # train_size = 0.7
                                                    ) 
```

```python
# 문제와 답이 갯수 동일한지 확인
# 꼭 .shape 사용해 확인 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 모델링
from sklearn.tree import DecisionTreeClassifier
```

```python
# 모델 객체 만들기
tree1 = DecisionTreeClassifier()
```

```python
# 모델 학습
tree1.fit(X_train, y_train)
```

```python
# 점수 체크해보기
tree1.score(X_test, y_test)
```

```python
# 트리모델 시각화
# 시각화 도구 불러오기
from sklearn.tree import plot_tree

# matplotlib 설정
plt.figure(figsize = (20,10), dpi =150)

# 트리 시각화
plot_tree(tree1, # 모델 지정
          filled = True, # 색상을 채워줘 -> 색상이 짙을수로 지니 불순도0에 가깝다
          feature_names= X_oh.columns, # 특성이름 표시하기
          class_names= y_train.unique() # 정답 이름 표시하기
         )
plt.show()
```

```python
# 트리모델 시각화
# 시각화 도구 불러오기
from sklearn.tree import plot_tree

# matplotlib 설정
plt.figure(figsize = (20,10), dpi =150)

# 트리 시각화
plot_tree(tree1, # 모델 지정
          filled = True, # 색상을 채워줘 -> 색상이 짙을수로 지니 불순도0에 가깝다
          feature_names= X_oh.columns, # 특성이름 표시하기, 질문의 이름이 인덱스번호로 나옴
          class_names= ['e', 'p'] # 정답 이름 표시하기
         )
plt.show()
```

```python
# 현재 그래프를 살펴 봤을때 Root Node(뿌리노드)에 항상 odor_n이 나오는 것을 볼 수 있다.
# 이유가 무엇일까?
# 특성 중요도를 살펴보고 정답 예측에 많은 영향을 미치는 컬럼을 살펴보자
fi = tree1.feature_importances_  # 지수 표현방식
fi
```

```python
# 데이터프레임으로 만들기
fi_df = pd.DataFrame(fi,
                    index = X_train.columns,
                    columns = ['특성중요도'])
fi_df
```

```python
# 위에서 만든 데이터 프레임을 내림차순 정렬해서 특성 중요도가 높은것부터 보자
fi_df.sort_values(by = '특성중요도', ascending = False)
# 특성중요도가 66퍼센트 나왔다 -> 정답을 예측하는데 가장 빠르게
# 특성중요도 높은순서부터 분기가 이루어진다.
# 특성중요도 0인것은 보지않음
```
